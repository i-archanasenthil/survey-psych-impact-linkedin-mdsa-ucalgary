##**Introduction**

LinkedIn has become an essential tool for showcasing skills, networking, and job hunting, playing a crucial role in shaping career opportunities for students entering the workforce. While it offers numerous advantages, such as increased visibility to employers and professional connections, its influence on mental well-being remains a growing concern. Research indicates that 72% of job seekers experience stress and imposter syndrome (Robinson, 2024; Mottram, 2023), suggesting that professional networking platforms may contribute to negative psychological effects.

This study aims to explore the relationship between LinkedIn usage patterns and students' mental health, focusing on self-confidence, imposter syndrome, motivation, and stress. By analyzing survey responses from MDSA students specializing in Data Science, Health Data Science, Financial Data Science, and Business Analytics, this research seeks to uncover whether frequent LinkedIn use exacerbates job-search anxiety or provides a sense of career progress and motivation.

Understanding these relationships is essential, as LinkedIn has become a standard platform in modern job searching. The findings of this study can offer valuable insights for students to manage LinkedIn usage effectively while minimizing its psychological downsides. Additionally, career advisors and educators can use this information to develop strategies that support students’ mental well-being during their job search, helping them leverage LinkedIn as a tool for success rather than a source of stress.

###**Test 1: Spearman Rank Test**

The Spearman Rank Test (Spearman's Rank Correlation Coefficient) is used to measure the relationship between ordinal variables. A significance level of 0.05 is applied to determine whether the correlation between the two ordinal variables is statistically significant.

```python
from scipy import stats

def perform_hypothesis_test(data, var1, var2):
    filtered_data = data[[var1, var2]].dropna()

    if filtered_data[var1].nunique() < 2 or filtered_data[var2].nunique() < 2:
        print(f"Spearman Correlation between {var1} and {var2}: Undefined (constant values)")
        return

    spearman_corr, p_value = stats.spearmanr(filtered_data[var1], filtered_data[var2])

    print(f"Spearman Correlation between {var1} and {var2}: {spearman_corr:.3f}, p-value: {p_value:.5e}")  # Scientific notation for small p-values

    if p_value < 0.05:
        print("Reject Null Hypothesis: There is a significant relationship.")
    else:
        print("Fail to Reject Null Hypothesis: No significant relationship found.")
    print("-" * 60)
```

####**1: Is there a correlation between positive interactions on LinkedIn (e.g., endorsements and messages) and self-confidence?**

**Null Hypothesis:** Positive Interactions and Self-Confidence are not correlated.

**Alternate Hypothesis:** Positive Interactions and Self-Confidence are correlated.

```python
# Running Hypothsesis for Do positive interactions on LinkedIn (e.g., endorsements and messages) influence self-confidence?

hypothesis_1=[("positive_interactions", "self_confidence")]
for var1, var2 in hypothesis_1:
    perform_hypothesis_test(survey, var1, var2)
```

### **Test 2: Score Test for Independence**

Because the ordinal values are in the categorical form, the chi-square test is used to check for association. A significance level of 0.05 is used to determine if the two ordinal values are independent of each other.

```python
def score_test(data, var1, var2):
    # Convert to binary categorical variables (1 = High, 0 = Low)
    data[f"{var1}_binary"] = (data[var1] >= 4).astype(int)
    data[f"{var2}_binary"] = (data[var2] >= 4).astype(int)

    # Create a contingency table
    table = pd.crosstab(data[f"{var1}_binary"], data[f"{var2}_binary"])

    # Perform Chi-Square test
    chi2, p, _, _ = stats.chi2_contingency(table)

    # Print Results
    print(f"Chi-square Test between {var1} and {var2}: χ² = {chi2:.3f}, p-value = {p:.5f}")
    if p < 0.05:
        print("Reject Null Hypothesis: Significant association exists.")
    else:
        print("Fail to Reject Null Hypothesis: No significant association found.")
    print("-" * 60)
```

#### **1: Is there any dependency between positive inteaction and self confidence?**

**Null Hypothesis:** Positive Interactions and Self Confidence are independent.

**Alternate Hypothesis:** Positive Interactions and Self Confidence are not independent.

```python
# H1: Do positive interactions boost confidence?

hypotheses_1 = [("positive_interactions", "self_confidence")]

for var1, var2 in hypotheses_1:
    score_test(survey, var1, var2)
```

Since the p-value is greater than alpha of 0.05, we fail to reject the null hypothesis. Thus we conclude that Positive Interactions and Self Confidence are independent of each other.

Since the p-value is greater than alpha of 0.05, we fail to reject the null hypothesis. Thus we conclude that Comparing oneself with feeling imposter syndrome are **independent**.

### **Test 3: Proportion Estimation**

The proportion estimation compares the proportion between two independent groups to determine if they are significantly different. This test is done to determine if the difference between proportion is different from 0.50

```python
import statsmodels.api as sm

def ratio_estimation(success_count, total_count, confidence=0.95):
    """Estimates proportion and confidence interval using Wilson's method."""
    prop = success_count / total_count
    conf_int = sm.stats.proportion_confint(success_count, total_count, alpha=1-confidence, method='wilson')

    print(f"Estimated proportion: {prop:.3f}, {confidence*100}% CI: {conf_int}")
    print("-" * 60)
```

####**1: Is the difference in proportion between positive inteaction and self confidence 0.05?**

**Null Hypothesis:** The difference in proportions between positive interaction and self-confidence is exactly 0.05.

**Alternate Hypothesis:** The difference in proportions between positive interaction and self-confidence is not equal to 0.05.

```python
ratio_tests_1 = [("comparing", "imposter_syndrome")]

# Perform Ratio Estimation
for var1, var2 in ratio_tests_1:
    high_var1 = (survey[var1] >= 4)
    high_var2 = (survey[var2] >= 4)

    success_count = (high_var1 & high_var2).sum()
    total_count = high_var1.sum()

    if total_count > 0:
        ratio_estimation(success_count, total_count)
    else:
        print(f"Not enough data for {var1} vs {var2}.")
```

We fail to reject the null hypothesis at the 5% significance level because the 95% confidence interval (0.3538, 0.8483) includes 0.50. This suggests that while our estimate (77.8%) indicates a potential effect, the **result is not statistically significant to reject null hypothesis**.

###**Test 4: LikeliHood Test**

Research studies indicate that 72% of job seekers experience a negative impact on their mental health. To assess whether this proportion holds true for the target population, a likelihood ratio test is conducted using our sample data.

**Null Hypothesis:** Proportion of people experiencing negative impact during job search is equal to 0.72

**Alternate Hypothesis:** Proportion of people experiencing negative impact during job search is different from 0.72

```python
count= survey.value_counts("imposter_syndrome")
count
```

```python
actual_responses = 27
agree_nege_effects = 14
# Expected was 72%
expected_negative_effects = round(0.72 * actual_responses)


# Observed frequencies: [agree, disagree]
observed_freq = [agree_nege_effects, actual_responses - agree_nege_effects]

# Expected frequencies: [agree, disagree]
expected_freq = [expected_negative_effects, actual_responses - expected_negative_effects]

# Perform chi-square test
chi2_stat, p_value = stats.chisquare(observed_freq, expected_freq)

print(f"Chi-square statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")
```

Since the p-value is less than 0.05, we **fail to reject** the null hypothesis. Therefore, we conclude that the proportion of people experiencing a negative impact on their mental well-being during a job search **differs from the reported 72%** in our sample.


###**Decision Tree**

A basic Decision Tree model was started as a baseline before tuning hyperparameters to optimize performance. To refine the model, GridSearchCV was implemented to test various hyperparameter combinations, including:

**Spearman Test**

The Spearman correlation analysis revealed that most variable pairs in our dataset do not exhibit statistically significant relationships. Positive interactions and self-confidence (ρ = 0.083, p = 0.6813), profile improvement and motivation in job search (ρ = 0.278, p = 0.1607), and job search tool effectiveness and feeling overwhelmed (ρ = -0.023, p = 0.9079) all failed to show meaningful correlations, as their p-values exceeded 0.05. However, a moderate positive correlation was found between self-comparison and imposter syndrome (ρ = 0.457, p = 0.01667), suggesting that individuals who compare themselves more with others tend to experience greater levels of imposter syndrome.

**Score Test**

The score test further supports the findings from the correlation analysis, confirming that the examined variables are independent of each other. Since all p-values were greater than 0.05, we failed to reject the null hypothesis in every case, indicating that positive interactions and self-confidence, self-comparison and imposter syndrome, profile improvement and job search motivation, and job search tool effectiveness and feeling overwhelmed do not have a statistically significant dependence. These results reinforce the idea that most factors studied do not exhibit strong relationships in our dataset.

**Proportion Estimation**

The proportion estimation analysis also failed to provide statistically significant evidence for any of the tested proportions. In all cases, the 95% confidence intervals included 0.50, leading to the failure to reject the null hypothesis at the 5% significance level. While our estimates suggested potential effects—such as 77.8% for two cases, 64.3%, and 26.7% for others—these findings were not strong enough to confirm statistical significance. This suggests that while there may be some observed trends, they are not robust enough to draw definitive conclusions.

The **decision tree** analysis revealed that challenges faced during job hunting was the most influential factor in classification, significantly impacting how students perceive LinkedIn's role in their job search.


These insights highlight the need for a balanced approach to LinkedIn usage, where students can leverage the platform for career growth while managing its psychological impacts.
